```
import numpy as np
import pandas as pd
```

```
pd.options.plotting.backend = "plotly"
```

```
x = np.linspace(-10, 10, 100)
z = 1/(1 + np.exp(-x))
z_derivative = z * (1 - z)
```

# Sigmoid Function

```
df = pd.DataFrame(dict(
    x = x,
    z = z
))
fig = df.plot.scatter(x="x", y="z")
fig.show()
```

# Derivative of the Sigmoid Function

```
df = pd.DataFrame(dict(
    x = x,
    z = z,
    z_derivative = z_derivative
))
fig = df.plot.scatter(x="x", y="z_derivative")
fig.show()
```

# Sigmoid Function with Derivative of the Sigmoid Function

```
df = pd.DataFrame(dict(
    x = x,
    z = z,
    z_derivative = z_derivative
))
fig = df.plot.scatter(x="x", y=["z_derivative", "z"])
fig.show()
```
