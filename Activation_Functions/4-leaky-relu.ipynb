```
import numpy as np
import pandas as pd
```

```
pd.options.plotting.backend = "plotly"
```

# Leaky ReLU

```
x = np.linspace(-10, 10, 100)
```

```
def ReLU(x):
    return x * (x > 0)
```

```
def LeakyReLU(z, alpha):
	return np.where(x > 0, x, x * alpha)
```

```
df = pd.DataFrame(dict(
    x = x,
    leakyrelu = LeakyReLU(x, 0.01),
    relu = ReLU(x)
))
```

```
fig = df.plot.scatter(x="x", y=["leakyrelu", "relu"])
fig.show()
```

# Derivative of Leaky ReLU

```
def dReLU(x):
    return 1 * (x > 0)
```

```
def dLeakyReLU(x, alpha):
    return np.where(x > 0, 1, alpha)
```

```
df = pd.DataFrame(dict(
    x = x,
    leakyrelu = LeakyReLU(x, 0.01),
    relu = ReLU(x),
    relu_derivative = dReLU(x),
    leaky_relu_derivative = dLeakyReLU(x, 0.01)
))
```

```
fig = df.plot.scatter(x="x", y=["relu_derivative", "leaky_relu_derivative"])
fig.show()
```
