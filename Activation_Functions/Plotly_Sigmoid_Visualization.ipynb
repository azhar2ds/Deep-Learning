**1. Sigmoid Activation function visualization**

```python
x = np.linspace(-10, 10, 100)
z = 1/(1 + np.exp(-x))
z_derivative = z * (1 - z)
df = pd.DataFrame(dict(x = x, z = z ))
fig = df.plot.scatter(x="x", y="z")
fig.show()
```

**2. Sigmoid Function**

```python
a = np.linspace(-3,3,100)
b = 1/(1 + np.exp(-x))

data=pd.DataFrame(dict(a=a, b=b))
fig=data.plot.scatter(x='a',y='b')
fig.show()

```

**3. Derivative of Sigmoid function**

```python
x = np.linspace(-10,10,100)
d = 1 / (1 + np.exp(-x)) 
dosf = d * (1-d)
data=pd.DataFrame(dict(x=x, dosf=dosf))
fig = data.plot.scatter(x='x', y='dosf')
fig.show()
```

**4. Sigmoid Function with Derivative of the Sigmoid Function**

```python
x = np.linspace(-10,10,200)
s = 1 / (1 + np.exp(-x))
d = s * (1-s)
dframe=pd.DataFrame(dict(x=x, s=s, d=d))
fig = dframe.plot.scatter(x='x', y=['s', d])
fig.show()
```

**7. Finding the sigmoid agian**

```python
x = np.linspace(-10,10,300)
s = 1 / (1 + np.exp(-x)) # finding the sigmoid
d = s * ( 1- s) # derivative of sigmoid function
dat=pd.DataFrame(dict(x = x, s= s, d=d))
fig=dat.plot.scatter(x='x', y=['s', d]) 
fig.show()
```
